```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://www.soumu.metro.tokyo.lg.jp/01soumu/archives/0702c_hensen_s1.htm#S2"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib')

table = soup.find('div', attrs = {'id':'enkaku'})


df={}
t=2
for Year in table.findAll('div',attrs={'class':'sc'}):
    #Dataframe for year level data
    Time={}
    
    #Insert data into year level dataframe
    Time['Year']=t
    Time['Admin']=list(filter(None,Year.findAll('div')[0].text.split(sep="\n")))
    Time['Dept']={}
    for i in range(1,len(Year.findAll('div'))-2):
        dept={}
        dept[Year.findAll('h4')[i-1].text]=list(filter(None,Year.findAll('div')[i].text.split(sep="\n")))
        Time['Dept'].update(dept)
    
    df[str(t)]=Time
    t=t+1

df=pd.DataFrame.from_dict(df).transpose()
```

```python
def Expand(df,time):    
    dta=pd.DataFrame()
    df=df[df["Year"]==time]
    df=pd.concat([df.drop(['Dept'],axis=1),df['Dept'].apply(pd.Series)],axis=1)
    print(df)
    for column in list(df.columns[1:]):
        Office=pd.DataFrame(df[column])
        Office["Dept"]=column
        Office.columns.values[0]="Office"
        Office=Office.explode("Office")


        dta = dta.append(Office, ignore_index = True)
    dta["Year"]=time
    return dta
```
