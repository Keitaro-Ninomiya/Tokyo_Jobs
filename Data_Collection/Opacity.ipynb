```python
from PIL import Image
import os, cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import re
import pytesseract
import pandas as pd

def get_concat_h(im1, im2):
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst

def listdir_nohidden1(path):
    for f in os.listdir(path):
        if not f.startswith('.'):
            yield f
            
class Convert_to_line:
    def __init__(self,im):
        self.im=im
    
    def get_concat_h(self,im1, im2):
        dst = Image.new('RGB', (im1.width + im2.width, im1.height))
        dst.paste(im1, (0, 0))
        dst.paste(im2, (im1.width, 0))
        return dst

    def crop(self):
        #Top Crop
        width,height=self.im.size
        top=0.3*height
        bottom=0.7*height

        left=0.1*width
        right=1*width
        IM1=self.im.crop((left,top,right,bottom))
        return IM1


from PIL import Image
 
def convertImage(image_test):
    img = image_test.convert("RGBA")
 
    datas = img.getdata()
 
    newData = []
 
    for item in datas:
        if item[0] > 150 and item[1] > 150 and item[2] > 150 :
            newData.append((255, 255, 255, 255))
        else:
            newData.append(item)
 
    img.putdata(newData)
    img.save("./New.png", "PNG")
    print("Successful")
    return img
```
