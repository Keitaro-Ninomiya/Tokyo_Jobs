```python
import json
import os
import pandas as pd
import numpy as np
import cv2


#このJson fileってなんだったっけ？1934の目次を読み取ったやつ
file_path =r'C:\Users\yuyaa.LAPTOP-UEQT9ACJ\Box\Tokyo_Jobs\Processed_Data\DataFrame.json'
with open(file_path, encoding="utf-8") as f:
    dta = json.loads(f.read())
```

```python
dta
```

```python
Year='1934'
Showa='9'

df = pd.read_csv(r'C:\Users\yuyaa.LAPTOP-UEQT9ACJ\Box\Tokyo_Jobs\Processed_Data\Index\S'+Showa+'.csv')

df=df.drop(df.columns[0], axis=1)
```

```python
df
```

```python
#ここのnってなんだっけ？
n=2
Row  = df.iloc[n]
Dept=Row['Dept']
Office=Row['Office']

Info=dta[str(Year)][Dept][Office]
PageList=Info['Page_Range']
print(Info)
StrPage=PageList[0]
StrCord=Info['Office_X1']
EndPage=PageList[-1]
EndCord=Info['Office_X2']

path='C:\\Users\\Keitaro Ninomiya\\Box\\Research Notes (keitaro2@illinois.edu)\\Tokyo_Jobs\\Raw_Data\\Splited\\'+str(Year)+'\\'
save_path='C:\\Users\\Keitaro Ninomiya\\Box\\Research Notes (keitaro2@illinois.edu)\\Tokyo_Jobs\\Raw_Data\\Office_Level\\'+str(Year)+'\\'+Dept+'\\'+Office+'\\'
os.chdir(save_path)

Count=1
if StrPage!=EndPage:
    for Page in PageList:
        if Page==StrPage:
            print("Starting at"+str(Page)+"|Count is"+str(Count))
            img=cv2.imread(path+"Page"+"{:03d}".format(Page)+"\\"+"Page"+"{:03d}".format(Page)+".jpg")
            HH, WW = img.shape[:2]
            img=img[0:HH,0:StrCord]
            cv2.imshow("Start",img)
            cv2.waitKey(0)
            cv2.imwrite('Image'+str(Count)+'.png', img)
            Count=Count+1
        if Page==EndPage:
            print("Ending at"+str(Page)+"|Count is"+str(Count))
            img=cv2.imread(path+"Page"+"{:03d}".format(Page)+"\\"+"Page"+"{:03d}".format(Page)+".jpg")
            HH, WW = img.shape[:2]
            img=img[0:HH,EndCord:WW]
            cv2.imshow("End",img)
            cv2.waitKey(0)
            cv2.imwrite('Image'+str(Count)+'.png', img)
        if (Page!=StrPage) and (Page<EndPage):
            print("Continuing at"+str(Page)+"|Count is"+str(Count))
            img=cv2.imread(path+"Page"+"{:03d}".format(Page)+"\\"+"Page"+"{:03d}".format(Page)+".jpg")
            HH, WW = img.shape[:2]
            img=img[0:HH,0:WW]
            cv2.imshow("Interim",img)
            cv2.waitKey(0)
            cv2.imwrite('Image'+str(Count)+'.png', img)
            Count=Count+1
if StrPage==EndPage:
    img=cv2.imread(path+"Page"+"{:03d}".format(Page)+"\\"+"Page"+"{:03d}".format(Page)+".jpg")
    HH, WW = img.shape[:2]
    img=img[0:HH,EndCord:StrCord]
    cv2.imwrite('Image'+str(Count)+'.png', img)
```
